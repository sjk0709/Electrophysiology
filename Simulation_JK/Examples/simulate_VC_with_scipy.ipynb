```python
from Models.kylie2017IKr import Kylie2017IKr
'''
Kylie 2017 IKr
'''
protocol = LeakStaircase()
model = Kylie2017IKr(protocol)
model.name = "Kylie 2017 IKr"
times = np.arange(77000)*0.0002  # times = np.linspace(0, 15.4, 77000)

V = model.voltage(times)
g = 0.1524 * 1e3 # [pA/V]
p1 = 2.26e-4 * 1e3 # [1/s]
p2 = 0.0699 * 1e3  # [1/V]
p3 = 3.45e-5 * 1e3 # [1/s]
p4 = 0.05462 * 1e3 # [1/V]
p5 = 0.0873 * 1e3  # [1/s]
p6 = 8.91e-3 * 1e3 # [1/V]
p7 = 5.15e-3 * 1e3 # [1/s]
p8 = 0.03158 * 1e3 # [1/V]

start_time = time.time()
simulator = scipy_simulator.Simulator(model)
simulator.model.set_params(g, p1, p2, p3, p4, p5, p6, p7, p8)
simulator.simulate(times)  
population_by_continent = {
    'O': model.open*model.active,
    'I': model.open*(1-model.active),
    'C': (1-model.open)*model.active,
    'IC': (1-model.open)*(1-model.active),    
}
print("--- %s seconds ---"%(time.time()-start_time))

plot_cell.plot_1D( model.times, model.V*1000, title='Staircase Protocol', figsize=(15,2), xlabel='Time (s)', ylabel='Voltage (mV)', label=None, xlim=(0, times.max()), ylim=None, save_path=None)
plot_cell.plot_1D( model.times, model.IKr*1000, title=None, figsize=(15,2), xlabel='Time (s)', ylabel='Current (pA)', label=None, xlim=(0, times.max()), save_path=None)
plot_cell.plot_stack( model.times, population_by_continent.values(), labels=population_by_continent.keys(), figsize=(15,2), xlabel='Time (s)', xlim=(0, times.max()), save_path=None )
```
