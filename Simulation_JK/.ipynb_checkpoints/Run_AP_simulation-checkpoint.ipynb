```python
from scipy.integrate import ode, solve_ivp
import os, sys, time
import numpy as np
import matplotlib.pyplot as plt
import pickle
import bisect

sys.path.append('./Protocols')
from pacing_protocol import PacingProtocol
from leakstaircase import LeakStaircase

sys.path.append('../')
import plot_cell
from simulator import Simulator
```

```python
from models.hh1592d import HH1592d
'''
Hodgkin Huxley 1592d
'''
protocol = PacingProtocol(level=1, start=5, length=0.03, period=30, multiplier=0, default_time_unit='ms')
model = HH1592d(protocol)
simulator = Simulator(model)
start_time = time.time()
model.name = "Hodgkin Huxley 1592d"
times = np.linspace(0,60,1000)
simulator.simulate(times, method='LSODA', max_step=8e-4)    
plot_cell.plot_AP(model.name, model.times, model.V)
print("--- %s seconds ---"%(time.time()-start_time))
```

```python
from models.br1977 import BR1977
'''
Beeler and Reuter 1977
'''
protocol = PacingProtocol(level=1, start=100, length=2, period=1000, multiplier=0, default_time_unit='ms')
model = BR1977(protocol)
simulator = Simulator(model)
start_time = time.time()
model.name = "Beeler and Reuter 1977"
times = np.linspace(0, 2000, 5000)
simulator.simulate(times)    
plot_cell.plot_AP(model.name, model.times, model.V)
print("--- %s seconds ---"%(time.time()-start_time))
```

```python
from models.kylie2017IKr import Kylie2017IKr
'''
Kylie 2017 IKr
'''
protocol = LeakStaircase()
model = Kylie2017IKr(protocol)
simulator = Simulator(model)
start_time = time.time()
model.name = "Kylie 2017 IKr"
times = np.arange(77000)*0.0002  # times = np.linspace(0, 15.4, 77000)
V = model.voltage(times)
g = 0.1524 * 1e3 # [pA/V]
p1 = 2.26e-4 * 1e3 # [1/s]
p2 = 0.0699 * 1e3  # [1/V]
p3 = 3.45e-5 * 1e3 # [1/s]
p4 = 0.05462 * 1e3 # [1/V]
p5 = 0.0873 * 1e3  # [1/s]
p6 = 8.91e-3 * 1e3 # [1/V]
p7 = 5.15e-3 * 1e3 # [1/s]
p8 = 0.03158 * 1e3 # [1/V]
simulator.model.set_params(g, p1, p2, p3, p4, p5, p6, p7, p8)
simulator.simulate(times)  
fig, ax = plt.subplots(figsize=(15,2))    
# fig.suptitle('New ORD-hERG - '+ ord._cell_types[cell_mode] +' cell', fontsize=16)
# ax.set_title('Simulation %d'%(simulationNo))
plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')     
ax.plot(model.times, model.V, label='Voltage')    
ax.legend()
plt.show()
# fig.savefig(os.path.join(result_folder, "voltage.jpg"), dpi=100)

fig, ax = plt.subplots(figsize=(15,2))    
# fig.suptitle('New ORD-hERG - '+ ord._cell_types[cell_mode] +' cell', fontsize=16)
# ax.set_title('Simulation %d'%(simulationNo))
plt.xlabel('Time (s)')
plt.ylabel('Current (pA)')     
ax.plot(model.times, model.IKr, label='IKr')    
# ax.set_xlim(0,16)
ax.legend()
plt.show()

population_by_continent = {
    'O': model.open*model.active,
    'I': model.open*(1-model.active),
    'C': (1-model.open)*model.active,
    'IC': (1-model.open)*(1-model.active),    
}
fig, ax = plt.subplots(figsize=(15,2))    
# fig.suptitle('New ORD-hERG - '+ ord._cell_types[cell_mode] +' cell', fontsize=16)
# ax.set_title('Simulation %d'%(simulationNo))
plt.xlabel('Time (s)')
# plt.ylabel('Voltage (mV)')     
ax.stackplot(model.times, population_by_continent.values(),
             labels=population_by_continent.keys(), alpha=0.8)
ax.legend()
plt.show()
# fig.savefig(os.path.join(result_folder, "Open_Active.jpg"), dpi=100)
print("--- %s seconds ---"%(time.time()-start_time))
```
