```python
from scipy.integrate import ode, solve_ivp
import os, sys, time
import numpy as np
import matplotlib.pyplot as plt
import pickle
import bisect

sys.path.append('../')
from Protocols.leakstaircase import LeakStaircase

sys.path.append('../')
import scipy_simulator
```

```python
from Models.kylie2017IKr import Kylie2017IKr
'''
Kylie 2017 IKr
'''
protocol = LeakStaircase()
model = Kylie2017IKr(protocol)
model.name = "Kylie 2017 IKr"
times = np.arange(77000)*0.0002  # times = np.linspace(0, 15.4, 77000)
t_eval = times
t_span = (0, times.max())

# V = model.voltage(times)
g = 0.1524 * 1e3 # [pA/V]
p1 = 2.26e-4 * 1e3 # [1/s]
p2 = 0.0699 * 1e3  # [1/V]
p3 = 3.45e-5 * 1e3 # [1/s]
p4 = 0.05462 * 1e3 # [1/V]
p5 = 0.0873 * 1e3  # [1/s]
p6 = 8.91e-3 * 1e3 # [1/V]
p7 = 5.15e-3 * 1e3 # [1/s]
p8 = 0.03158 * 1e3 # [1/V]

start_time = time.time()
simulator = scipy_simulator.Simulator(model)
simulator.model.set_params(g, p1, p2, p3, p4, p5, p6, p7, p8)
simulator.simulate(t_span=t_span, t_eval=t_eval, method='LSODA', max_step=8e-4)  
population_by_continent = {
    'O': model.open*model.active,
    'I': model.open*(1-model.active),
    'C': (1-model.open)*model.active,
    'IC': (1-model.open)*(1-model.active),    
}
print("--- %s seconds ---"%(time.time()-start_time))
```
