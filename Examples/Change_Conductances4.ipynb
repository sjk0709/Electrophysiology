```python
import os, sys, time, copy
import numpy as np
import matplotlib.pyplot as plt

import myokit
sys.path.append('../')
sys.path.append('../Protocols')
sys.path.append('../Models')
sys.path.append('../Lib')
import protocol_lib
import mod_trace

import simulator_myokit
import simulator_scipy
```

```python
end_time = 1000
bcl = 1000
duration = 0.5
offset = 20

celltype = 2

t_span = (0, end_time)
t_eval = np.linspace(0, t_span[1], 5000)  
```

```python
start_time = time.time()

model_path = "../mmt-model-files/ohara-cipa-v1-2017.mmt" 
model, p, s = myokit.load(model_path)
protocol = myokit.pacing.blocktrain(bcl*100, duration, offset=offset) # period, duration, offset=0, level=1.0, limit=0  
sim = simulator_myokit.Simulator(model, protocol, max_step=1.0, abs_tol=1e-06, rel_tol=1e-6, vhold=0)  # 1e-12, 1e-14  # 1e-08, 1e-10
sim.name = "ORD2017"  
sim.simulation.set_constant('cell.mode', celltype)    
y0 = sim.pre_simulate(bcl*100, sim_type=1)
d = sim.simulate(end_time, log_times=None, extra_log=['ina.INa', 'ikr.IKr', 'iks.IKs', 'ito.Ito', 'ical.ICaL', 'ik1.IK1', 'inal.INaL'])

sol1 = {}
times1 = d['engine.time']
sol1["Voltage"] = d['membrane.V']
sol1["I_Na"] = d['ina.INa']
sol1["I_Kr"] = d['ikr.IKr']
sol1["I_Ks"] = d['iks.IKs']
sol1["I_To"] = d['ito.Ito']
sol1["I_CaL"] = d['ical.ICaL']
sol1["I_K1"] = d['ik1.IK1']
sol1["I_NaL"] = d['inal.INaL']
I_total = np.zeros_like(times1)
for i, (name, values) in enumerate(sol1.items()):   
    if i != 0:
        I_total += values
sol1["I_total"] = I_total

print("--- %s seconds ---"%(time.time()-start_time))
```

```python
start_time = time.time()

model_path = "../mmt-model-files/ohara-cipa-v1-2017_JK-v1.mmt" 
model, p, s = myokit.load(model_path)
protocol = myokit.pacing.blocktrain(bcl*100, duration, offset=offset) # period, duration, offset=0, level=1.0, limit=0  
sim = simulator_myokit.Simulator(model, protocol, max_step=1.0, abs_tol=1e-06, rel_tol=1e-6, vhold=0)  # 1e-12, 1e-14  # 1e-08, 1e-10
sim.name = "ORD2017"  

print("--- %s seconds ---"%(time.time()-start_time))
```

```python
start_time = time.time()

sim.simulation.set_constant('setting.simType', 0)
sim.simulation.set_constant('cell.mode', celltype)       
       
params = {         
    'ina.gNa' : 75.0,   
    'inal.gNaL' : 0.0075 * 2.661,  
    'ito.gto' : 0.02,
    'ical.PCa' : 0.0001 * 1.007,
    'ikr.gKr' : 4.65854545454545618e-2, # [mS/uF]
    'iks.gKs' : 0.0034 * 1.87,
    'ik1.gK1' : 0.1908 * 1.698,
    'inaca.gNaCa' : 0.0008,
    'inak.PNaK' : 30,
    'ikb.gKb' : 0.003,
    'inab.PNab' : 3.75e-10,
    'icab.PCab' : 2.5e-8,
    'ipca.GpCa' : 0.0005,

    'ina.g_adj' : 1,  
    'inal.g_adj' : 1,
    'ito.g_adj' : 4,
    'ical.g_adj' : 2.5,
    'ikr.g_adj' : 0.8,
    'iks.g_adj' : 1,
    'ik1.g_adj' : 1.3,
    'inaca.g_adj' : 1.4,
    'inak.g_adj' : 0.7,
    'ikb.g_adj' : 1,
    'inab.g_adj' : 1,
    'icab.g_adj' : 1,
    'ipca.g_adj' : 1, 
}
sim.set_simulation_params(params)

y0 = sim.pre_simulate(bcl*100, sim_type=1)
d = sim.simulate(end_time, log_times=None, extra_log=['membrane.V', 'ina.INa', 'ikr.IKr', 'iks.IKs', 'ito.Ito', 'ical.ICaL', 'ik1.IK1', 'inal.INaL'])

sol2 = {}
times2 = d['engine.time']
sol2["Voltage"] = d['membrane.V']
sol2["I_Na"] = d['ina.INa']
sol2["I_Kr"] = d['ikr.IKr']
sol2["I_Ks"] = d['iks.IKs']
sol2["I_To"] = d['ito.Ito']
sol2["I_CaL"] = d['ical.ICaL']
sol2["I_K1"] = d['ik1.IK1']
sol2["I_NaL"] = d['inal.INaL']
I_total = np.zeros_like(times2)
for i, (name, values) in enumerate(sol2.items()):   
    if i != 0:
        I_total += values
sol2["I_total"] = I_total

print("--- %s seconds ---"%(time.time()-start_time))
```
